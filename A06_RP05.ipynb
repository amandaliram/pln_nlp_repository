# **Roteiro de Práticas 05:** Representação Numérica 2 - A Matemática da Relevância

Este laboratório prático foca na evolução do modelo *Bag-of-Words* para o **TF-IDF**. O objetivo é aprender a penalizar palavras muito comuns (como "o", "de", "e") e dar destaque aos termos que realmente carregam o significado estatístico de um documento.

### **1. Exemplo 1: Briefing do Problema (O Cenário)**
  Em um sistema de *Social Listening*, nem todas as palavras têm o mesmo peso. Se um cliente diz "O produto é excelente", a palavra "excelente" é muito mais informativa para o negócio do que a palavra "O". O **TF-IDF** resolve isso multiplicando a frequência local do termo (TF) pela raridade global do termo no corpus (IDF).

**1.1 Dataset de Trabalho:**

```python
#Dataset de Trabalho:
corpus = [
    "o processamento de linguagem natural é incrível",
    "linguagem natural e inteligência artificial",
    "o processamento de linguagem e inteligência",
    "artificial incrível e natural",
    "inteligência é incrível"
]
```

**1.2 Módulo 1: Entendendo o Componente TF (Term Frequency)**   
O **TF** mede a frequência de um termo em um documento específico. Ele responde: "Quão importante é esta palavra neste texto?".

1.   Use o `CountVectorizer` para gerar a contagem bruta de termos.
2.   Observe como palavras como "linguagem" ou "incrível" aparecem com pesos iguais a conectivos se usarmos apenas contagem simples.




```python
# Módulo 1: Entendendo o Componente TF (Term Frequency)
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

vectorizer = CountVectorizer()
tf_matrix = vectorizer.fit_transform(corpus)

# Visualizando como DataFrame para análise
df_tf = pd.DataFrame(tf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
print("Matriz de Frequência Bruta (TF):")
display(df_tf)
```

**1.3 Módulo 2: O Peso do Mundo (Inverse Document Frequency - IDF)**   
O **IDF** é o "filtro de raridade". Se uma palavra aparece em todos os documentos, seu IDF será baixo. Se aparece em apenas um, seu IDF será alto. O scikit-learn utiliza uma versão suavizada da fórmula do IDF:
Onde *n* é o total de documentos e *df(t)* é o número de documentos que contêm o termo *t*.

1.   Instancie o `TfidfVectorizer`.
2.   Após o fit, acesse o atributo .idf_ para ver o peso de cada palavra.

```python
# Módulo 2: O Peso do Mundo (Inverse Document Frequency - IDF)
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)

# Mapeando termos aos seus pesos IDF
df_idf = pd.DataFrame(tfidf_vectorizer.idf_, index=tfidf_vectorizer.get_feature_names_out(), columns=["Peso IDF"])
print("Pesos IDF (Quanto maior, mais rara/importante é a palavra no contexto global):")
display(df_idf.sort_values(by="Peso IDF", ascending=False))
```

**1.4 Módulo 3: Construindo a Matriz TF-IDF Final**   
A matriz final é o produto de *TF x IDF*. O resultado é uma representação vetorial onde os termos irrelevantes são "achatados" em direção ao zero. Converta a matriz esparsa final em um `DataFrame` e compare o peso da palavra "natural" (que aparece em 3 frases) com "artificial" (que aparece em 2).

```python
# Módulo 3: Construindo a Matriz TF-IDF Final

df_tfidf_final = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("Matriz TF-IDF Final (Normalizada):")
display(df_tfidf_final)
```

**3.3 Etapa 3 – Criação das coleção de documentos**

```python
# Função para ler e copiar o conteúdo de um arquivo (documento)
def ler(nome_arquivo):
    arquivo = open(nome_arquivo,'r', encoding='iso-8859-1') # codificacao latin-1
    conteudo_arq = arquivo.read()
    arquivo.close()
    return conteudo_arq
```

```python
# Função para fazer a limpeza do texto
def limpar(lista):
    lixo = '.,:;?!"``^~()[]{}/\|@#$%¨&*-'
    quase_limpo = [x.strip(lixo).lower() for x in lista]
    return [x for x in quase_limpo if x.isalpha() or '-' not in x]
```

```python
# Geração dos endereços dos arquivos
obras = []

for i in range(1,10):
    obras.append('/root/nltk_data/corpora/machado/romance/marm0' + str(i) +'.txt')

print(obras)
```

```python
colecao = []
for obra in obras:
    texto = ler(obra)
    print(texto)
    palavras = limpar(texto.lower().split())
    colecao.append(palavras)
```

**3.4 Etapa 4 – Cálculo do TF-IDF**

```python
import math

def tf(termo,doc):
    return colecao[doc].count(termo) / len(colecao[doc])

def df(termo):
    return len([d for d in colecao if termo in d])

def idf(termo):
    return math.log10(len(colecao) / df(termo))

def tf_idf(termo, doc):
    return tf(termo, doc) * idf(termo)
```

```python
tf('capitu',7)
# tf(termo,documento)
```

```python
idf('capitu')
# tf(termo)
```

```python
tf_idf('capitu',7)
# tf(termo,documento)
```

```python
# Função para listagem de ordenação pela relevância das palavras
def mais_relevantes(doc):
    lista_total = [(tf_idf(p, doc), p)for p in set(colecao[doc])]

    return sorted(lista_total, reverse=True)[:50]
```

# 🏆 Projeto Final (PBL): Social Listening com TF-IDF

**Objetivo:** Evoluir o nosso modelo de Bag-of-Words (BoW) para o **TF-IDF**.
Enquanto o BoW apenas conta a frequência bruta das palavras, o TF-IDF atua como um "filtro de raridade". Ele penaliza conectivos comuns (como "the", "and", "is") e dá peso matemático aos termos que realmente descrevem o contexto das avaliações de celulares da Amazon, revelando os reais sentimentos e problemas dos clientes.

```python
# ==============================================================================
# PREPARAÇÃO TOTAL: Imports, Download, Limpeza e TF-IDF
# ==============================================================================
# 1. Instalação e Importações Essenciais
!pip install kagglehub -q
import pandas as pd
import numpy as np
import re
import nltk
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer
import kagglehub
import glob
import os

nltk.download('punkt_tab', quiet=True)

# 2. Recriando o nosso Tokenizer Profissional (O motivo do seu erro!)
def tokenizer_personalizado(texto):
    if not isinstance(texto, str):
        return []
    texto = texto.lower()
    texto = re.sub(r'<.*?>', ' ', texto)
    texto = re.sub(r'http\S+', '', texto)
    texto = re.sub(r'[^\w\s]', ' ', texto)
    return word_tokenize(texto)

# 3. Baixando e Carregando o Dataset da Amazon
print("📦 Baixando o dataset do Kaggle e preparando os dados...")
path = kagglehub.dataset_download("abdallahwagih/amazon-reviews")
arquivos_json = glob.glob(os.path.join(path, '**', '*.json'), recursive=True)
arquivo_json = arquivos_json[0]

# Lendo o JSON e isolando as 15.000 amostras
df_amazon = pd.read_json(arquivo_json, lines=True)
colunas_texto = df_amazon.select_dtypes(include=['object']).columns
coluna_alvo = 'reviewText' if 'reviewText' in df_amazon.columns else ('body' if 'body' in df_amazon.columns else ('review' if 'review' in df_amazon.columns else colunas_texto[0]))
corpus_amazon = df_amazon[coluna_alvo].dropna().astype(str).tolist()[:15000]

# 4. A Aplicação do TF-IDF (O passo que você tentou dar)
print("⏳ Iniciando a vetorização com TF-IDF (Isso levará alguns segundos)...")

tfidf_vectorizer = TfidfVectorizer(tokenizer=tokenizer_personalizado, token_pattern=None)
matriz_tfidf = tfidf_vectorizer.fit_transform(corpus_amazon)

print("="*60)
print(f"✅ TF-IDF CONCLUÍDO COM SUCESSO!")
print(f"📊 A nova matriz possui {matriz_tfidf.shape[0]} reviews e {matriz_tfidf.shape[1]} termos únicos.")
print("="*60)
```

### 🎯 Critério de Aceite: As 10 Palavras Mais Relevantes
Para descobrir sobre o que os clientes da Amazon mais falam, vamos somar o "peso" TF-IDF de cada termo ao longo de todo o dataset. As palavras com maior pontuação final serão os temas centrais das avaliações.

```python
# ==============================================================================
# PASSO 2: Extração de Features (A Magia do TF-IDF)
# ==============================================================================
# Somamos os pesos TF-IDF de cada coluna (palavra) através de todos os documentos
soma_tfidf = matriz_tfidf.sum(axis=0)

# Extraímos os nomes das palavras e os conectamos aos seus pesos totais
palavras = tfidf_vectorizer.get_feature_names_out()
pesos = np.squeeze(np.asarray(soma_tfidf))

# Criamos um DataFrame do Pandas para facilitar a ordenação
df_relevancia = pd.DataFrame({'Palavra': palavras, 'Peso_TF_IDF_Total': pesos})

# Ordenamos do maior para o menor peso e pegamos as Top 10
top_10_palavras = df_relevancia.sort_values(by='Peso_TF_IDF_Total', ascending=False).head(10)

print("="*60)
print("🏆 AS 10 PALAVRAS MAIS RELEVANTES DO DATASET AMAZON (TF-IDF)")
print("="*60)
display(top_10_palavras)
```

### 🔍 O TF-IDF em Ação (Zoom em uma Review)
Vamos pegar uma avaliação aleatória para provar como o algoritmo ignora palavras comuns e dá peso apenas aos termos que definem o problema ou o elogio do cliente.

```python
# ==============================================================================
# PASSO 3: O TF-IDF em Ação (Zoom em uma Review Específica)
# ==============================================================================
# Escolhendo a review de índice 42 (pode testar com outros números!)
id_review = 42
review_escolhida = corpus_amazon[id_review]

# Extraindo os pesos TF-IDF APENAS desta review específica
pesos_review = matriz_tfidf[id_review].toarray().flatten()
df_review = pd.DataFrame({'Palavra': palavras, 'Peso_TF_IDF': pesos_review})

# Filtrando apenas as palavras que apareceram na frase e ordenando pelo peso
termos_destaque = df_review[df_review['Peso_TF_IDF'] > 0].sort_values(by='Peso_TF_IDF', ascending=False)

print(f"📝 REVIEW BRUTA DO CLIENTE:\n'{review_escolhida}'\n")
print("🔍 Como a IA avaliou a importância de cada palavra nesta frase:")
display(termos_destaque.head(5))
```

### 💡 Reflexão Técnica: TF-IDF vs. Bag-of-Words no E-commerce

**Por que o TF-IDF é superior ao Bag-of-Words (BoW) simples para classificar se um comentário é uma reclamação técnica ou um elogio genérico?**

A superioridade do TF-IDF está na sua capacidade de atuar como um **filtro de relevância e raridade**, corrigindo a "ingenuidade" do BoW.

1. **A Falha do BoW:** O Bag-of-Words apenas conta frequências. Em um e-commerce, palavras genéricas como *"produto"*, *"comprei"*, *"chegou"*, *"muito"* e *"bom"* aparecem aos milhares, tanto em elogios quanto em reclamações. O BoW daria um peso gigantesco a essas palavras, afogando o algoritmo em ruído.
2. **A Solução do TF-IDF:** O TF-IDF resolve isso penalizando matematicamente os termos que aparecem em quase todos os documentos (reduzindo o peso de palavras genéricas do dia a dia) e multiplicando o peso de palavras raras que aparecem concentradas em poucos documentos.
3. **Na Prática (Reclamações Técnicas):** Um elogio genérico usa um vocabulário muito comum. Já uma reclamação técnica possui um vocabulário específico (ex: *"bateria"*, *"esquentando"*, *"trincado"*, *"defeito"*, *"reembolso"*). O TF-IDF identifica que essas palavras são raras no escopo global (IDF alto), mas muito frequentes naquela avaliação específica (TF alto), dando a elas um peso altíssimo.

**Conclusão:** O TF-IDF permite que o algoritmo de Machine Learning "ignore" o vocabulário padrão do e-commerce e foque apenas nos termos específicos que caracterizam uma dor real do cliente, tornando a classificação de sentimentos muito mais precisa e barata computacionalmente.
