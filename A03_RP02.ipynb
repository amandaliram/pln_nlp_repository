```python
texto_exemplo = "Olá, mundo! Como vai você?"
print(f"Texto original: '{texto_exemplo}'")
```

```python
# A função .split() quebra a string em uma lista de strings
tokens = texto_exemplo.split()
print(f"Tokens (divididos por espaço): {tokens}")
print(f"Quantidade: {len(tokens)}")
```

```python
texto_com_pontuacao = "Olá, mundo! Como vai você?"
tokens_com_pontuacao = []
token_atual = ""

# 1.5 Percorremos cada caractere do texto
for caractere in texto_com_pontuacao:
    # Verificamos se o caractere é uma letra ou um número
    if caractere.isalnum():
        token_atual += caractere #Adiciona o caractere ao token atual
    else:
        # Se o caractere não for alfanumérico, é uma pontuação ou espaço
        if token_atual:
            tokens_com_pontuacao.append(token_atual)
            token_atual = ""
        # Adiciona a pontuação como um token separado, se não for um espaço
        if caractere.isspace() == False:
            tokens_com_pontuacao.append(caractere)

# Adiciona o último token após o loop
if token_atual:
    tokens_com_pontuacao.append(token_atual)

print(f"Tokens com pontuação separada: {tokens_com_pontuacao}")
print(f"Quantidade: {len(tokens_com_pontuacao)}")
```

### **2. Exemplo 2: Tokenização Manual**

```python
import unicodedata

texto_acentuado = "Vocês já pensaram na água do Oceano? É fascinante!"
print(f"Texto original: '{texto_acentuado}'")
```

```python
# Normaliza a string para a forma 'NFD' (Normalized Form Decomposition)
# que separa o caractere base (ex: 'e') do acento (ex: '´')
texto_normalizado = unicodedata.normalize('NFD', texto_acentuado)

# Filtra a string para manter apenas os caracteres sem acento
texto_sem_acentos = ''.join(
    [char for char in texto_normalizado if char.isalnum() or char.isspace()]
)

print(f"Texto sem acentos: '{texto_sem_acentos}'")
```

```python
import re

# Adicione esta função para tokenizar e limpar o texto de forma completa
def tokenizar_e_limpar(texto):
    # 1. Converte o texto para minúsculas
    texto = texto.lower()

    # 2. Normaliza para remover acentos
    texto_normalizado = unicodedata.normalize('NFD', texto)

    # 3. Usa uma expressão regular para remover pontuação e caracteres não-alfanuméricos
    # O [^\w\s] remove qualquer coisa que não seja letra, número ou espaço
    texto_limpo = re.sub(r'[^\w\s]', '', texto_normalizado, re.UNICODE)

    # 4. Quebra o texto em tokens (palavras)
    tokens = texto_limpo.split()

    return tokens

# Usamos a função para processar o texto original
tokens_finais = tokenizar_e_limpar(texto_acentuado)

print(f"Tokens finais (limpos e sem acentos): {tokens_finais}")
```

### **3. Preparando o Terreno: O Módulo re**

  O Python possui um módulo nativo chamado re (Regular Expressions). Ele é o seu bisturi para operar o texto bruto.


**3.1 Exemplo 3: Identificando Padrões Simples**  
Crie uma célula no Google Colab e execute o código abaixo para entender como substituir ruídos (como pontuação excessiva) por espaços.

```python
import re

# Exemplo de comentário sujo de e-commerce
comentario_bruto = "O produto é excelente!!!!!! Mas a entrega... atrasou. :( "

# Removendo pontuação excessiva com Regex
# O padrão [^\w\s] significa: "pegue tudo que NÃO for letra, número ou espaço"
comentario_limpo = re.sub(r'[^\w\s]', '', comentario_bruto)

print(f"Bruto: {comentario_bruto}")
print(f"Limpo: {comentario_limpo}")
```

**3.2 Exemplo 4: Removendo Lixo da Web**  
Em diversos conjuntos textuais, os dados vêm carregados de links e marcações HTML que poluem a análise estatística.  Use o padrão abaixo para limpar URLs e tags HTML de uma só vez.

```python
def faxina_web(texto):
    # Remove tags HTML (ex: <p>)
    texto = re.sub(r'<.*?>', '', texto)
    # Remove URLs (http/https)
    texto = re.sub(r'http\S+', '', texto)
    return texto.strip()

print(faxina_web("<p>Confira a oferta em: https://loja.com/promo</p>"))
```

Relatório de Impacto: Calcule o tamanho do vocabulário (Types) antes e depois da limpeza.

```python
import re
import nltk
from nltk.tokenize import word_tokenize

# Garante que o pacote de tokenização do NLTK está disponível
nltk.download('punkt_tab', quiet=True)
print("Bibliotecas importadas com sucesso!")
```

```python
# ========================================================================
# ETAPA 1: Escolha do texto
# Requisito: Pelo menos três frases, com símbolos, e caracteres variados.
# ========================================================================
texto_desafio = """
<h1>Atraso na entrega do pedido #98765!</h1>
Estou muito decepcionado com a compra, o produto não chegou no prazo estipulado... 😡
Vocês podem verificar o status em https://lojaexemplo.com.br/rastreio?id=98765.
Aguardo um retorno urgente, pois era um presente de aniversário (para minha mãe)!!!
"""
print("Texto bruto carregado na memória!")
```

```python
# ========================================================================
# ETAPA 2: Implementação do Pipeline
# ========================================================================
def pipeline_limpeza(texto):
    # A. Conversão para minúsculas.
    texto = texto.lower()

    # B. Remoção de tags HTML e URLs.
    texto = re.sub(r'<.*?>', ' ', texto)
    texto = re.sub(r'http\S+', '', texto)

    # C. Remoção de caracteres especiais (mantendo apenas o texto/alfanuméricos).
    texto = re.sub(r'[^\w\s]', ' ', texto)

    # D. Tokenização via NLTK.
    tokens = word_tokenize(texto)

    return tokens

print("Função 'pipeline_limpeza' criada e pronta para uso!")
```

```python
# ========================================================================
# ETAPA 3: Processamento e Cálculo do Vocabulário (Types)
# ========================================================================

# Calculando os Types ANTES da limpeza (usando split básico para simular a sujeira)
tokens_brutos = texto_desafio.split()
types_antes = len(set(tokens_brutos))

# Aplicando o pipeline e calculando os Types DEPOIS da limpeza
tokens_limpos = pipeline_limpeza(texto_desafio)
types_depois = len(set(tokens_limpos))

print(f"Processamento concluído. Foram gerados {len(tokens_limpos)} tokens limpos.")
```

```python
# ========================================================================
# ETAPA 4: Relatório de Impacto e Justificativa de Negócios
# ========================================================================
print("="*60)
print("🎯 RELATÓRIO DE IMPACTO - DESAFIO E-COMMERCE")
print("="*60)
print(f"📊 Vocabulário (Types) ANTES do pipeline: {types_antes}")
print(f"📊 Vocabulário (Types) DEPOIS do pipeline: {types_depois}")
print(f"📉 Redução de ruído no vocabulário: {types_antes - types_depois} types (lixo removido).")
print("-" * 60)
print("💡 JUSTIFICATIVA PARA O CLIENTE (Otimização em Nuvem):")
print("A redução de Types (palavras únicas) diminui a dimensionalidade da nossa base")
print("de dados. Em um ambiente de Cloud Computing (como AWS, Azure ou GCP), o custo")
print("é diretamente baseado no consumo de memória RAM e no tempo de processamento.")
print("Ao removermos 'lixos' como URLs únicas, tags HTML e pontuações grudadas nas")
print("palavras, nossos modelos de Processamento de Linguagem Natural vão lidar com")
print("matrizes matemáticas muito menores, rodando de forma mais rápida, eficiente")
print("e significativamente mais barata para o orçamento da empresa.")
print("="*60)
```
