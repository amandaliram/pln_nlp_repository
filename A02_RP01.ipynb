**1.1 Números (Inteiros e Fracionários)**

**1.2 Sequências (Strings, Listas e Tuplas)**

````python
# Tarefa 1.2: Manipulando Strings.

# Strings são sequências de caracteres
frase = "A linguística computacional é fascinante."
print(f"A frase inteira: {frase}")

# Acesso por índice (o índice começa em 0)
print(f"Primeiro caractere: {frase[0]}")
print(f"Último caractere: {frase[-1]}")

# Acessando um trecho (slice)
print(f"Primeiras 12 letras: {frase[:12]}")

# Métodos de strings
print(f"Frase em maiúsculas: {frase.upper()}")
print(f"Frase em minúsculas: {frase.lower()}")
````

### **3. Funções e Filtragem de Dados**

  Agora, vamos combinar tudo o que aprendemos para realizar uma tarefa prática: filtrar dados.


**3.1 Funções Definidas pelo Programador**

````python
# Tarefa 3.1: Crie uma função nomeada.

# Usamos 'def' para definir uma função. Ela recebe um 'argumento'
def contar_palavras(sentenca):
    """Conta o número de palavras em uma sentença."""
    # A função processa o argumento e retorna um valor
    palavras = sentenca.split()  # Converte a string em uma lista de palavras
    return len(palavras)

# Chamar a função com um texto
sentenca_exemplo = "O rato roeu a roupa do rei de Roma"
numero_de_palavras = contar_palavras(sentenca_exemplo)
print(f"O número de palavras na frase é: {numero_de_palavras}")
````

````python
# Tarefa 3.2: Entenda as Funções Anônimas (lambda).

# Funções lambda são pequenas funções de uma linha.
# Elas são úteis para tarefas rápidas.
multiplicar = lambda x, y: x * y

resultado = multiplicar(5, 3)
print(f"O resultado da multiplicação é: {resultado}")
````

**3.2 Filtragem de Dados (Integrando tudo)**  
Vamos agora usar tudo o que aprendemos para filtrar uma lista de palavras.

````python
# Tarefa 3.3: Filtre palavras que começam com a letra 'a'.

texto_completo = "A análise de sentimentos é uma das áreas mais aplicadas do PLN."
# Converta o texto para minúsculas e divida em uma lista de palavras
tokens_sujos = texto_completo.lower().split()
print(f"Tokens originais: {tokens_sujos}")

# Vamos criar uma lista para guardar o resultado
palavras_filtradas = []

# Use um loop 'for' e uma condicional 'if' para filtrar
for palavra in tokens_sujos:
    if palavra.startswith('a'):
        palavras_filtradas.append(palavra)

print(f"Palavras que começam com 'a': {palavras_filtradas}")
````

````python
# Tarefa 3.4: O Desafio Pythonico - Use a Compreensão de Listas.
# A mesma tarefa anterior pode ser feita de forma mais elegante e
# concisa em Python usando uma Compreensão de Listas.

# Mesmo resultado, em uma única linha!
palavras_filtradas_compreensao = [palavra for palavra in tokens_sujos if palavra.startswith('a')]
print(f"Resultado com Compreensão de Listas: {palavras_filtradas_compreensao}")
````

**1. Escreva uma função que receba uma string e retorne uma nova string correspondente à original invertida (de trás para frente). Para isso, crie um loop com while. O resultado deve ser parecido com este:**
```
>>> inversa('abc')
'cba'
```

````python
def inversa(texto):
    resultado = ""
    # Começamos pelo último índice da string
    indice = len(texto) - 1

    # O loop continua até chegarmos ao primeiro caractere (índice 0)
    while indice >= 0:
        resultado += texto[indice]
        indice -= 1

    return resultado

# Testando o resultado esperado no documento
print(inversa('abc'))
````

**2. Escreva uma função que receba uma sentença e retorne o número (inteiro) de palavras que ela contém.**
```
>>> palavras('O rato roeu a roupa do rei de Roma')
9
```



````python
def palavras(sentenca):
    # O método .split() divide a string nos espaços, criando uma lista.
    # A função len() conta quantos itens existem nessa lista.
    lista_de_palavras = sentenca.split()
    return len(lista_de_palavras)

# Testando o resultado esperado no documento
print(palavras('O rato roeu a roupa do rei de Roma'))
````

**3. Escreva uma função que receba uma sentença e retorne uma lista das letras iniciais de cada palavra em maiúsculas.**
```
>>> iniciais('O rato roeu a roupa do rei de Roma')
['O', 'R', 'R', 'A', 'R', 'D', 'R', 'D', 'R']
```



````python
def iniciais(sentenca):
    # Dividimos a sentença e pegamos o índice [0] (primeira letra) de cada palavra.
    # O .upper() garante que fique em maiúscula.
    return [palavra[0].upper() for palavra in sentenca.split()]

# Testando o resultado esperado no documento
print(iniciais('O rato roeu a roupa do rei de Roma'))
````

**4. Escreva uma função que receba uma string e retorne o conjunto das letras que ela contém, ou seja, o “alfabeto” que a produziu. Dica: você vai querer eliminar o espaço em branco do conjunto. É mais fácil fazer isso depois de gerar o conjunto.**
```
>>> alfabeto('O rato roeu a roupa do rei de Roma')
{'I', 'R', 'O', 'A', 'D', 'M', 'E', 'P', 'U', 'T'}
```



````python
def alfabeto(sentenca):
    # Converte tudo para maiúsculas e transforma em um conjunto (elimina duplicatas)
    letras_unicas = set(sentenca.upper())

    # Verifica se há espaço em branco no conjunto e o remove
    if ' ' in letras_unicas:
        letras_unicas.remove(' ')

    return letras_unicas

# Testando o resultado esperado no documento
print(alfabeto('O rato roeu a roupa do rei de Roma'))
````
